# Cheapest Cost and Highest earning States

```python
scorecard['md_earn_wne_p10'] = scorecard['md_earn_wne_p10'].apply(lambda x: np.nan if x == "PrivacySuppressed" else float(x))
```

```python
save = scorecard.groupby('STABBR').mean()
```

```python
cols = ['NPT4_PRIV', 'NUM4_PRIV']
scorecard[cols] = scorecard[cols].fillna(0)
```

```python
cost_by_state = save[['NPT4_PUB', 'NPT4_PRIV', 'NUM4_PUB', 'NUM4_PRIV', 'md_earn_wne_p10']].sort_values(["NPT4_PUB", "NPT4_PRIV", "NUM4_PUB", "NUM4_PRIV", "md_earn_wne_p10"], axis=0, ascending=False)
cost_by_state
```

```python
def weighted_cost(row):
    return (row['NPT4_PUB'] * row['NUM4_PUB'] + row['NPT4_PRIV'] * row['NUM4_PRIV'])/ (row['NUM4_PUB'] + row['NUM4_PRIV'])

cost_by_state['avg_cost'] = cost_by_state.apply(weighted_cost, axis=1).sort_values(ascending=False)
cost_by_state
```

```python
highest_earning = cost_by_state['md_earn_wne_p10']/cost_by_state['avg_cost']
highest_earning.sort_values()
# NH is the worst option whereas HI is the best with highest earning
```

# Models: Logistic Regression 

```python
from sklearn import linear_model
from sklearn.ensemble import RandomForestClassifier
```

```python
columns1 = ['NPT4_PUB', 'NPT4_PRIV', 'NPT4_PROG', 'NPT4_OTHER','md_earn_wne_p10']
scorecard[columns1] = scorecard[columns1].fillna(0)
```

```python
lm = linear_model.LinearRegression()

X = scorecard[['NPT4_PUB', 'NPT4_PRIV', 'NPT4_PROG', 'NPT4_OTHER']].values
y = scorecard['md_earn_wne_p10']

model = lm.fit(X, y)
```

```python
predictions  =  model.predict(X)
score        =  model.score(X, y)

# Plot the model
plt.figure(figsize=(8,8))
plt.scatter(predictions, y, s=30, c='r', marker='+', zorder=10)

plt.show()

print "score: ", score

# the predictions and the target values are way off because the points are all gather in one place like cloud form.
```
